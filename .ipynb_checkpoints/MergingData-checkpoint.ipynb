```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Se realiza la concatenación de los saber pro desde 2016 a 2021

separador = '¬'
base_path = '/Users/santi/ModeloP2/txt/SaberPro/'

años = [2016, 2017, 2018, 2019, 2020, 2021]
dataframes = []

for año in años:
    if año == 2021:
        archivo = f'SaberPro_{año}.txt'
    else:
        archivo = f'SaberPro_Genéricas_{año}.txt' 
    ruta = base_path + archivo
    df = pd.read_csv(ruta, sep=separador, engine='python', quotechar="'")
    dataframes.append(df)

df_saberPro = pd.concat(dataframes, ignore_index=True)

df_saberPro
```

```python
df_cruces = pd.read_csv('/Users/santi/ModeloP2/txt/Cruces/Llave_Saber11_SaberPro.txt', sep=separador, engine='python')
df_cruces
```

```python
df_cruces.info()
```

```python
#separo el identificador de saber11 y saberPro

df_cruces[['ESTU_CONSECUTIVO_saber11', 'ESTU_CONSECUTIVO_saberPro']] = df_cruces['id_master,id_using'].str.split(',', expand=True)

df_cruces
```

```python
# Hago merge del dataframe de cruces y el de saber11
df_saber11 = df_cruces.merge(
    df_saber11, left_on='ESTU_CONSECUTIVO_saber11', right_on='ESTU_CONSECUTIVO', how='inner')

df_saber11
```

```python
# Hago merge del dataframe de cruces y el de saberPro
df_saberPro = df_cruces.merge(
    df_saberPro, left_on='ESTU_CONSECUTIVO_saberPro', right_on='ESTU_CONSECUTIVO', how='inner')

df_saberPro
```

```python
# Merge final
merged_df = df_saber11.merge(df_saberPro, on='ESTU_CONSECUTIVO_saber11', how='inner')

merged_df
```
