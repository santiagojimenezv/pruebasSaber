```python
# Define la función para asignar el área de conocimiento
def asignar_area(programa_academico, areas_conocimiento):
    for area, programas in areas_conocimiento.items():
        if programa_academico in programas:
            return area
    return "Desconocido"

# Agrega la columna 'Area_Basica_Conocimiento' al DataFrame original
df_mzls['Area_Basica_Conocimiento'] = df_mzls['ESTU_PRGM_ACADEMICO'].apply(lambda x: asignar_area(x, areas_conocimiento))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from tabulate import tabulate

# Seleccionar las columnas relevantes para el modelo
selected_columns = ['PUNT_MATEMATICAS', 'PUNT_INGLES', 'PUNT_LECTURA_CRITICA', 
                     'COLE_NOMBRE_ESTABLECIMIENTO', 'ESTU_GENERO_y', 'edad', 'ESTU_AREARESIDE_y', 'Area_Basica_Conocimiento']

df_model = df_mzls[selected_columns]

# Filtrar datos para el colegio 'COL. SAN LUIS GONZAGA'
colegio_seleccionado = 'ASPAEN-GIMNASIO LOS CEREZOS'
df_colegio = df_model[df_model['COLE_NOMBRE_ESTABLECIMIENTO'] == colegio_seleccionado]

# Eliminar filas con NaN en las puntuaciones
df_colegio.dropna(subset=['PUNT_MATEMATICAS', 'PUNT_INGLES', 'PUNT_LECTURA_CRITICA'], inplace=True)

# Definir las features (X) y las etiquetas (y)
X_columns = ['ESTU_GENERO_y', 'edad', 'ESTU_AREARESIDE_y', 'Area_Basica_Conocimiento']
y_columns = ['PUNT_MATEMATICAS', 'PUNT_INGLES', 'PUNT_LECTURA_CRITICA']

# Dividir el conjunto de datos para entrenamiento y prueba
train_data, test_data = train_test_split(df_colegio, test_size=0.2, random_state=42)

# Definir transformadores para preprocesamiento
numeric_features = ['edad']
categorical_features = ['ESTU_GENERO_y', 'ESTU_AREARESIDE_y', 'Area_Basica_Conocimiento']

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Crear el modelo con preprocesamiento y regresión aleatoria
model = Pipeline(steps=[('preprocessor', preprocessor),
                        ('regressor', RandomForestRegressor(random_state=42))])

# Definir features (X) y etiquetas (y) para el modelo
X_train = train_data[X_columns]
y_train = train_data[y_columns]
X_test = test_data[X_columns]
y_test = test_data[y_columns]

# Entrenar el modelo
model.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
predictions = model.predict(X_test)

# Evaluar el rendimiento del modelo
mse = mean_squared_error(y_test, predictions)
print(f"Error cuadrático medio para el colegio '{colegio_seleccionado}': {mse}")

# Ahora puedes usar el modelo entrenado para hacer predicciones en nuevos datos
areas_conocimiento = ["AGRONOMÍA, VETERINARIA Y AFINES", "BELLAS ARTES", "CIENCIAS DE LA EDUCACIÓN",
                       "CIENCIAS DE LA SALUD", "CIENCIAS SOCIALES Y HUMANAS", 
                       "ECONOMÍA, ADMINISTRACIÓN, CONTADURÍA Y AFINES", 
                       "INGENIERÍA, ARQUITECTURA, URBANISMO Y AFINES", 
                       "MATEMÁTICAS Y CIENCIAS NATURALES"]

new_data = pd.DataFrame({
    'ESTU_GENERO_y': ['F'] * len(areas_conocimiento),
    'edad': [23.0] * len(areas_conocimiento),
    'ESTU_AREARESIDE_y': ['Cabecera Municipal'] * len(areas_conocimiento),
    'Area_Basica_Conocimiento': areas_conocimiento
})

# Realizar predicciones
predictions_new = model.predict(new_data)

# Mostrar las predicciones en una tabla
predicted_scores = pd.DataFrame(predictions_new, columns=y_columns, index=areas_conocimiento)
table = tabulate(predicted_scores, headers='keys', tablefmt='fancy_grid', showindex=True)

print(f'Predicciones para el colegio "{colegio_seleccionado}":')
print(table)
```
